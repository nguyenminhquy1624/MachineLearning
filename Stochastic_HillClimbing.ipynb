```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective(x):
  return x[0] ** 2.0

def hillclimbing(objective, bounds, n_iteration, step_size):
  solution = bounds[:,0] + np.random.rand(len(bounds)) * (bounds[:,1] - bounds[:,0])
  solution_eval = objective(solution)
  scores = list()
  scores.append(solution_eval)
  for i in range(n_iteration):
    candidate = solution + np.random.randn(len(bounds)) * step_size
    candidate_eval = objective(candidate)
    if candidate_eval <= solution_eval:
      solution, solution_eval = candidate, candidate_eval
      scores.append(solution_eval)

  return [solution, solution_eval, scores]
```

```python
r_min, r_max = -5.0, 5.0

input = np.arange(r_min, r_max, 0.1)
result = [objective([x]) for x in input]

plt.plot(input, result)
x_optima = 0.0
plt.axvline(x = x_optima, ls = '--', color ='red')
plt.show()
```

```python
np.random.seed(5)
bounds = np.asarray([[-5.0,5.0]])
n_iter = 10000
step_size = 0.1
best, score, scores = hillclimbing(objective, bounds, n_iter, step_size)
print("Best = ", best)
print("Score = ", score)
```
