```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective(x):
  return x**2.0

def derivative(x):
  return x * 2.0

def gradient_decent(objective, derivative, bounds, n_iter, step_size):
  solution = bounds[:,0] + np.random.rand(len(bounds)) * (bounds[:,1] - bounds[:,0])

  solutions , scores = list(), list()

  for i in range(n_iter):
    gradient = derivative(solution)
    solution =solution - gradient*step_size
    solution_eval = objective(solution)

    solutions.append(solution)
    scores.append(solution_eval)

  return [solutions, scores]
```

```python
bounds = np.asarray([[-1.0,1.0]])
n_iter = 100
step_size = 0.1
solutions, scores = gradient_decent(objective, derivative, bounds, n_iter, step_size)

inputs = np.arange(bounds[0,0], bounds[0,1] + 0.1, 0.1)
results = objective(inputs)
plt.plot(inputs, results)
plt.plot(solutions, scores, '.-', color='red')
plt.show()
```
