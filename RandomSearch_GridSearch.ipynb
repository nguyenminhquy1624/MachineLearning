```python
import numpy as np
import matplotlib.pyplot as plt
```

# Tìm kiếm ngẫu nhiên để tối ưu hóa hàm

```python
def objective(x):
  return x**2.0

r_min, r_max = -5.0, 5.0
sample = r_min + np.random.rand(100)*(r_max - r_min)
sample_eval = objective(sample)

best_idx = 0
for i in range(len(sample)):
  if sample_eval[i] < sample_eval[best_idx]:
    best_idx = i

print("Best idx = ", sample[best_idx])
print("Sample Eval = ", sample_eval[best_idx])
```

```python
print("Sample = ", sample.shape)
print("Sample Eval = ", sample_eval.shape)
```

```python
inputs = np.arange(r_min, r_max, 0.1)
results = objective(inputs)
plt.plot(inputs, results)
plt.scatter(sample, sample_eval)
plt.axvline(x=sample[best_idx], ls= '--', color='red')
plt.show()
```

# Grid Search để tối ưu hóa hàm

```python
def objective(x,y):
  return x**2.0 + y**2.0

r_min, r_max = -5.0, 5.0
sample = list()
step = 0.3

for x in np.arange(r_min, r_max, step):
  for y in np.arange(r_min, r_max, step):
    sample.append([x,y])

sample_eval = [objective(x,y) for x, y in sample]

best_idx = 0

for i in range(len(sample)):
  if sample_eval[i] < sample_eval[best_idx]:
    best_idx = i
print("Sample best idx", sample[best_idx])
print("Sample eval best idx", sample_eval[best_idx])
```

```python
from numpy.lib.function_base import meshgrid
xaxis = np.arange(r_min, r_max, 0.1)
yaxis = np.arange(r_min, r_max, 0.1)
x, y = meshgrid(xaxis, yaxis)
result = objective(x,y)
plt.contourf(x,y,result, level=50, cmap='jet')
plt.plot([x for x,_ in sample], [y for _,y in sample], '.', color ='black')
plt.plot(sample[best_idx][0], sample[best_idx][1], '*', color='white')
plt.show()
```
