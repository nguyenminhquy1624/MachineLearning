# Optimizer Local Optima

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def objective(x, y):
 return -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2)))
 - np.exp(0.5 * (np.cos(2 * np.pi * x)
 + np.cos(2 * np.pi * y))) + np.e + 20
```

```python
r_min, r_max = -5.0, 5.0
xaxis = np.arange(r_min, r_max, 0.1)
yaxis = np.arange(r_min, r_max, 0.1)
print(xaxis.shape)
print(yaxis.shape)
```

```python
x, y = np.meshgrid(xaxis,yaxis)
result = objective(x,y)
axis = plt.subplot(projection='3d')
axis.plot_surface(x,y, result, cmap='jet')
plt.show()
```

```python
from scipy.optimize import basinhopping
```

```python
def objective(v):
 x, y = v
 return -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2 * np.pi * x) +
               np.cos(2 * np.pi * y))) + np.e + 20

pt = r_min + np.random.rand(2)*(r_max-r_min)
results = basinhopping(objective, pt, stepsize=0.5, niter=200)
solution = results['x']
evaluation = objective(solution)
print('Solution = ', solution)
print('Evaluation = ', evaluation)
```
